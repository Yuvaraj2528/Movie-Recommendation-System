```python
import numpy as np
import pandas as pd
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
movies_data=pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with 'null' string 
for x in selected_features:
    movies_data[x]=movies_data[x].fillna('')
```

```python
#combining all the selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
#Obtaining similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
print(similarity)
similarity.shape
```

```python
# getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

```python
#creating movies name list from the dataset
titles=movies_data['title'].tolist()
print(titles)
```

```python
movie_match = difflib.get_close_matches(movie_name, titles)
print(movie_match)
best_match=movie_match[0]
print(best_match)
```

```python
#getting index based of title of the movie
index_of_movie=movies_data[movies_data['title']==best_match]['index'].values[0]
index_of_movie
```

```python
#getting the similarity score from the index of movie 
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_scores = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
#print(sorted_scores)
sortt=sorted_scores[1:]
sortt
```

```python
# print the name of similar movies based on the index
print('Suggested Movies are:')
i=1
for x in sortt:
    index=x[0]
    suggested_title=movies_data[movies_data['index']==index]['title'].values[0]
    if(i<11):
        print(i,'. ',suggested_title)
        i+=1
```

# Final Movie System:

```python
movie_name = input(' Enter your favourite movie name : ')

titles=movies_data['title'].tolist()

movie_match = difflib.get_close_matches(movie_name, titles)

best_match=movie_match[0]

index_of_movie=movies_data[movies_data['title']==best_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_scores = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
final_sort=sorted_scores[1:]

print('Suggested Movies are:')
i=1
for x in final_sort:
    index=x[0]
    suggested_title=movies_data[movies_data['index']==index]['title'].values[0]
    if(i<11):
        print(i,'. ',suggested_title)
        i+=1
        
```
